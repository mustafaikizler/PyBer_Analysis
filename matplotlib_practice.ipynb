```python
#Import Dependencies.
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#create the plot, (chain the plot funsction to plt)

plt.plot(x_axis, y_axis)
```

```python
#create the plot

plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")



#create label for x and y axes,

plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title

plt.title("Pyber Frade by Month")

#add a legend

plt.legend()

#add a grid

plt.grid()

#Figure size




```

```python
#OBJECT ORIENTED APPROACH
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# HORIZONAL BAR CHART WIth object oriented approach

#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#SKILL DRILL

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
plt.gca().invert_yaxis()

ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_label("Chicago")
ax.legend()
ax.set_title("PyBer Fare by Month")
```

```python
# CREATE A BUBBLE CHART

plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Make the bubbles bigger by multiplying the Y_AXIS with any number (we will iterate in order to ready all the data)
```

```python
y_axis_larger= []

for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# List Compherension technique to use to iterate throug y_axis = MUCH MORE EASIER # BEAUTY OF PYTHON

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#SKILL DRILL 
#Using the object-oriented approach, make the following changes to your bubble chart:

fig, ax = plt.subplots()
ax.scatter (y_axis,x_axis , c = "b", s = [i * 5 for i in y_axis], alpha= 0.5,  edgecolors="black", linewidths=2, )
plt.gca().invert_yaxis()
ax.set_label("Boston")

ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

```

```python
#Assingn 12 colors, one for each month,
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]

plt.subplots (figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct="%.1f%%")

plt.show()
```

Create a Pie Chart Using the Object-Oriented Interface

```python
#SKILL DRILL
explodes =  [0,0,0.2,0,0,0,0.2,0,0,0,0,0]

fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis,labels = x_axis, autopct = '%.1f%%', explode = explodes, shadow= True, startangle=97, counterclock=False,  colors= colors)

 
plt.show()
```
