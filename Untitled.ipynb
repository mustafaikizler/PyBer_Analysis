```python
import matplotlib.pyplot as plt
import statistics

# We are adding statistics librarry in order to add y-errors to line and abr charts.
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# calculate the standard deviation of the y_axis

stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=4, capthick=4)
```

```python
#Object Oriented Appriach and Error Bars
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3, ecolor="red")
```

## Add Error Bars to a Bar Chart


```python
plt.bar(x_axis,y_axis, yerr = stdev, capsize = 3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

## Change the Major Ticks

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
np.arange(0,51, step = 5) 
```

```python
#Object oriented method

fg, ax = plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51, step = 4.5))
```

```python
#ADD MINOR TICKS
```

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure,

fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

#Create minor ticks at an increment  of 1.
plt.gca().invert_yaxis()
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()






```
